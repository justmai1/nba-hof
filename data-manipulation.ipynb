## Player Accolades

```python
all_star = pd.read_csv("data/nba-all-star.csv")
awards = pd.read_csv("data/nba-awards.csv")
end_season = pd.read_csv("data/nba-end-season-teams.csv")
longevity = pd.read_csv("data/nba-longevity.csv")
```

```python
all_star.groupby(["player"]).count().sort_values("season", ascending=False)
```

```python
awards.loc[awards["winner"] == True].groupby(
    ["player_id", "player", "award"]
).count()["season"]
```

```python
end_season.groupby(["player_id", "player", "type", "number_tm"]).count()["season"]
```

```python
longevity.groupby(["player_id", "player"]).max(
    "experience").sort_values(
        "experience", ascending=False)
```
